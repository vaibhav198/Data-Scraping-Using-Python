# Importing packages


```python
#Python module to fetch urls
import urllib.request
#Python library to pull the data from HTML and XML files
from bs4 import BeautifulSoup as BS
#data manipulation tool
import pandas as pd
#library for scientific computing
import numpy as np
```

```python
#url of the webpage from where we want to extract the data
url = "https://www.iiitm.ac.in/index.php/en/alumini/alumini-our-pride"
page = urllib.request.urlopen(url)
soup = BS(page)

print(soup.title.string)
```

```python
all_a_tags = soup.find_all("a")
all_tables = soup.find_all("table")
```

```python
l = []

for row in all_tables:
    columns = row.findAll("p")
    for c in columns:
        l.append(c.string.split("\n"))
```

```python
print(len(l))
for item in range(len(l)):
    print(l[item][0])
```

```python
Name_of_alumini = []
Designation = []
Company = []

#for i in range(int(len(list)/3) - 1):
#    A.append(i+1)
columns = ['Name of Alumini', 'Designation', 'Company']
for i in range(3, (len(l) -2), 3 ):
    Name_of_alumini.append(l[i][0])
    Designation.append(l[i+1][0])
    Company.append(l[i+2][0])
our_pride = pd.DataFrame(columns=columns)
our_pride['Name of Alumini'] = Name_of_alumini
our_pride['Designation'] = Designation
our_pride['Company']  = Company
our_pride
```
